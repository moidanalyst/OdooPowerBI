## Create Connection

```python
import xmlrpc.client
import pandas as pd

# Replace with your Odoo credentials
url = 'url_here'
db = 'db_name'
username = '_email'
password = 'your_password_here'

# Authenticate
common = xmlrpc.client.ServerProxy(f'{url}/xmlrpc/2/common')
uid = common.authenticate(db, username, password, {})
if not uid:
    raise Exception("Authentication failed!")

# Connect to the object proxy
models = xmlrpc.client.ServerProxy(f'{url}/xmlrpc/2/object')

# Test connection by listing available models
models_list = models.execute_kw(db, uid, password, 'res.partner', 'search_read', [[]], {'fields': ['name', 'email'], 'limit': 5})
# print(models_list)

# Storing models list in DataFrame
df = pd.DataFrame(models_list)
display(df)
```

## Retrieve Invoices

```python
invoices = models.execute_kw(
    db, uid, password,
    'account.move', 'search_read',  # Method to search and read records
    [[['move_type', '=', 'out_invoice']]],  # Filter for customer invoices
    {'fields': ['id', 'name', 'partner_id', 'amount_total', 'invoice_date'], 'limit': 5}  # Fields to retrieve
)

# Print retrieved invoices
# for invoice in invoices:
#     print(invoice)

df = pd.DataFrame(invoices)
display(df)
```

## Retrieve Payments

```python
fields = models.execute_kw(db, uid, password, 'account.payment', 'fields_get', [])

# for field, properties in fields.items():
#     print(f"Field: {field}")
#     print(f"Label: {properties['string']}")
#     print(f"Type: {properties['type']}")
#     print("--------")

payments = models.execute_kw(
    db, uid, password,
    'account.payment', 'search_read',
    [],  # Correct filter format
    {
        'fields': [
            'id', 'date', 'name', 'amount', 'payment_method_code', 'currency_id', 
            'partner_id', 'payment_reference', 'payment_type'
        ],
        'limit': 5
    }
)

df = pd.DataFrame(payments)
display(df)

# df.to_csv('fields.csv')

# Print retrieved payments
# for payment in payments:
#     print(payment)
```
